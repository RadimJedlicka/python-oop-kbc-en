## Class properties

---


In some programming languages, attributes are **not accessed directly**.

The reason is to prevent the user from unintentionally working with an incorrect data type or an invalid value.

In Python this functionality is **not supported directly**, but there is a *workaround*.

<br>

Imagine a situation where you need to write a **volume unit converter**.

From **litres** to **pints** (UK):

<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Ftse2.mm.bing.net%2Fth%3Fid%3DOIP.FCRrMRa9SQrNjgFSKfKP6AHaHa%26pid%3DApi&f=1&ipt=f45d961e5b9d1d428a1c59180cd22f9c6a298a6a8ccaa753d86ca54f906f045b&ipo=images" width="200" style="margin-left:auto; margin-right:auto"/>

````python
class VolumeConverter:
    def __init__(self, litre: int):
        self.litre = litre
        self.pint_coefficient = 1.7598
        self.pints = self.litre * self.pint_coefficient
````

````python
two_litres = VolumeConverter(2)
````

````python
two_litres.pints
````

<br>

As a user you can still pass an **incorrect argument** to your class:

````python
three_litres = VolumeConverter("three")
````

<br>

Or if you need to recalculate such an instance, the *interpreter* will not respond to your changes:

````python
two_litres.litre = 3
````

````python
two_litres.pints
````

````python
two_litres.litre
````

<br>

In this case, the *interpreter* works **with the first value** it received during *instantiation*.

Subsequent changes only affect the instance attribute `litre`, not the previously prepared `pints`.

### Solution using a method

---

Your first reaction might be to add a **suitable method**.

The situation is certainly manageable here, but you have to **break the existing code**.

In other words, everyone who works with such a script will have their code disrupted:

````python
class VolumeConverter:

    def __init__(self, litre: int):
        self.litre = litre
        self.pint_coefficient = 1.7598
        
    def pints(self) -> float:
        return self.litre * self.pint_coefficient
````

````python
two_litres = VolumeConverter(2)
````

````python
two_litres.pints()
````

<br>

Everywhere the user worked **with instances and attributes**, they will suddenly have to modify the code:

````python
two_litres.litre = 5
````

````python
two_litres.litre
````

````python
two_litres.pints()
````

<br>

The result is now correct.

But fixing existing code can be **challenging**. Especially if the class is not this simple.

<br>

### Solution using the `property` function

---

In other languages, users immediately reach for:
1. `setter`,
2. `getter`,
3. `deleter`.

That is, objects that allow proper manipulation and setting of attributes.

Python offers several tricks in this regard, first the `property` function:

````python
class VolumeConverter:

    def __init__(self, litre: int):
        self.litre = litre
        self.__pint_coefficient = 1.7598
        
    @property
    def pints(self):
        return self.litre * self.__pint_coefficient
````

<br>

In very simplified terms, you essentially create methods, but the *interpreter* allows you to access them as attributes.

````python
two_litres = VolumeConverter(2)
````

````python
two_litres.pints
````

````python
two_litres.pints()
````

<br>

In the current notation, you **no longer explicitly use the method and parentheses**.

But only access through the *attribute* `pints`.

````python
two_litres.litre = 3
````

````python
two_litres.pints
````

<br>

But what if you want to enter **pints directly** and convert them to litres:

````python
two_litres.pints = 2
````

<br>

The *interpreter* won't let you assign a value to the instance attribute.

<br>

### Setter, getter decorators

---

If you want to enter **litres separately** and **pints separately**, you need to properly structure your class:

````python
class VolumeConverter:

    def __init__(self):
        self._litre  = 0
        self._pint = 0
        self.__pint_coefficient = 1.7598
        
    @property
    def litre(self):
        print("GETTER: retrieving value 'litre'.")
        return self._litre
    
    @property
    def pint(self):
        print("GETTER: retrieving value 'pint'.")
        return self._pint
````

<br>

Using so-called *getters*, you prepare the attributes `pint` and `litre`, which you will now use every time you access the attribute:

````python
converter_1 = VolumeConverter()
````

````python
print(
    converter_1.litre,
    converter_1.pint,
    sep="\n"
)
````

<br>

**Getters** are only used for querying and need to be marked as protected variables (*weak private*): 

````python
converter_1.litre = 2
````

Now, however, you don't have a way to set values.

You can now **only retrieve** values.

<br>

To be able to conveniently change attribute values, you need to set the `setter` decorator:

````python
class VolumeConverter:

    def __init__(self):
        self._litre = 0
        self._pint = 0
        self.__pint_coefficient = 1.7598

    @property
    def litre(self):
        print("GETTER: retrieving value 'litre'.")   # FOR ILLUSTRATION ONLY
        return self._litre

    @property
    def pint(self):
        print("GETTER: retrieving value 'pint'.")  # FOR ILLUSTRATION ONLY
        return self._pint
    
    @litre.setter
    def litre(self, value: float):
        print("SETTER: setting value 'litre'.")  # FOR ILLUSTRATION ONLY
        self._litre = value  

    @pint.setter
    def pint(self, value: float):
        print("SETTER: setting value 'pint'.") # FOR ILLUSTRATION ONLY
        self._pint = value
````

````python
converter_1 = VolumeConverter()
````

````python
print(converter_1.litre)
````

````python
converter_1.litre = 2
````

````python
converter_1.pint = 2
````

````python
print(
    converter_1.litre,
    converter_1.pint,
    sep='\n'
)
````

<br>

In this case, Python first executes `converter_1.litre` and `converter_1.pint`. That's why you see:
```
GETTER: retrieving value 'litre'.
GETTER: retrieving value 'pint'.
```

Only later, the `print` function itself outputs the results:
```
2
2
```

<br>

Now that you can elegantly enter both pints and litres, you can add methods for conversions:

````python
class VolumeConverter:

    def __init__(self):
        self._litre = 0
        self._pint = 0
        self.__pint_coefficient = 1.7598
        
    @property
    def pint(self):
        print("GETTER: retrieving value 'pint'.")
        return self._pint
    
    @property
    def litre(self):
        print("GETTER: retrieving value 'litre'.")
        return self._litre
    
    @pint.setter
    def pint(self, value: float):
        print("SETTER: setting value 'pint'.")
        self._pint = value
        
    @litre.setter
    def litre(self, value: float):
        print("SETTER: setting value 'litre'.")
        self._litre = value
        
    def to_pints(self):
        if self._litre:
            return self.__pint_coefficient * self._litre   # round()
        else:
            print("Value for 'litre' not set")
    
    def to_litres(self):
        if self._pint:
            return self._pint / self.__pint_coefficient  # round()
        else:
            print("Value for 'pint' not set")
````

````python
converter_1 = VolumeConverter()
````

````python
converter_1.to_pints()
````

````python
converter_1.litre
````

````python
converter_1.litre = 3
````

````python
converter_1.litre
````

````python
converter_1.to_pints()
````

````python
converter_1.pint = 3
````

````python
converter_1.to_litres()
````

### 🧠 EXERCISE 🧠, Create a class `Area` that will consist of the following:
---

1. Define a class `Area`,
2. create an *instance constructor* that doesn't require parameters,
3. the *instance constructor* only defines a *weak private* attribute `self._diameter_cm` set to zero,
4. create a `getter` method (using a decorator) named `diameter_cm`,
5. create a `setter` method (using a decorator) that verifies the `value` parameter is either `int` or `float`, otherwise prints `"Invalid data type for 'diameter_cm'"`
6. create an *instance method* `calculate_circle_area` that uses the constant `math.pi` (from the `math` library) to return the area of a circle,
7. create a static method `calculate_circle_circumference` that uses the parameter `radius` to return the circumference of a circle,
8. create a magic method `__str__` that formats the output to string: `Circle with diameter: <radius>, area: <area>.`

````python
import math
````

````python
class Area:
    
    def __init__(self):
        self._diameter_cm = ""
    
    @property
    def diameter_cm(self):
        return self._diameter_cm
    
    @diameter_cm.setter
    def diameter_cm(self, value):
        if isinstance(value, int) or isinstance(value, float):  # (int, float, decimal.Decimal)
            self._diameter_cm = value
        else:
            # logging.warning()
            raise TypeError("Invalid data type for 'diameter_cm'")
    
    def calculate_circle_area(self):
        radius = self._diameter_cm / 2
        return math.pi * ((radius)**2)

    def calculate_circle_circumference(self, radius):
        return 2* math.pi * radius
    
    def __str__(self):
        return f"Circle with diameter: {self._diameter_cm}, area: {round(self.calculate_circle_area(), 3)}."
````

````python
a1 = Area()
````

````python
a1.diameter_cm = 10
````

````python
a1.diameter_cm
````

````python
a1.diameter_cm = "10"
````

````python
print(
    a1,
    str(a1),
    sep='\n'
)
````

3. define `@property` named `temp_celsius` with a corresponding *getter* that returns the value `self._temp_celsius`,
4. define `@property` named `temp_celsius` with a corresponding *setter* that checks the value is not lower than **-273.15 degrees Celsius** (otherwise `ValueError`),
5. define `@property` named `temp_fahrenheit` with a corresponding *getter* that returns the value `self._temp_fahrenheit`,
6. define `@property` named `temp_fahrenheit` with a corresponding *setter*,
7. define a **static method** `celsius_to_fahrenheit` that accepts temperature **in degrees Celsius** and returns the corresponding temperature **in degrees Fahrenheit**,
8. define a **static method** `fahrenheit_to_celsius` that accepts temperature **in degrees Fahrenheit** and returns the corresponding temperature **in degrees Celsius**.

````python
class TemperatureConverter:
    def __init__(self):
        self._temp_celsius = 0.0

    @property
    def temp_celsius(self):
        return self._temp_celsius

    @temp_celsius.setter
    def temp_celsius(self, value):
        # logger.info("Converting to degrees Celsius..")
        if value < -273.15:
            # logger.warning("...")
            raise ValueError("Temperature cannot be lower than -273.15 degrees Celsius.")
        self._temp_celsius = value
        # logger.info("Value stored as 'temp_celsius'")

    @property
    def temp_fahrenheit(self):
        return TemperatureConverter.celsius_to_fahrenheit(self._temp_celsius)  # 

    @temp_fahrenheit.setter
    def temp_fahrenheit(self, value):
        self._temp_celsius = TemperatureConverter.fahrenheit_to_celsius(value) # 

    @staticmethod
    def celsius_to_fahrenheit(temp_celsius):
        return temp_celsius * 9 / 5 + 32

    @staticmethod
    def fahrenheit_to_celsius(temp_fahrenheit):
        return (temp_fahrenheit - 32) * 5 / 9

    def __str__(self):
        return f"Temperature: {self._temp_celsius}°C, {self.temp_fahrenheit}°F"
````

---

### Terminology

---

**Polymorphism** means *many forms*.

In simple terms, you can imagine a situation where one thing can be described in different ways.

For example, a **greeting**. A polite greeting can be expressed in several ways, or even in different languages.

<br>

*Polymorphism* contributes to code flexibility and extensibility by allowing programmers to use **objects of different classes interchangeably**.

### Meaning in Python

---

In the world of programming, this means that one object can serve different purposes:

````python
a = 1
````

````python
print(a.__add__(9))
````

````python
print(
    1 + 1,              # operator '+' with integer operands,
    "Lesson" + " #03",  # operator '+' with string operands,
    sep="\n"
)
````

<br>

You already know this concept.

The `+` operator means **two different procedures for two different data types**:
1. `int`, addition,
2. `str`, concatenation.

<br>

However, you don't only know such polymorphism between `int` and `str`.

You can notice similar behavior with the built-in function `len`:

````python
print(
    len("Matous"),                                                         # str
    len(["city", "sea", "chicken", "building"]),                          # list
    len({"name": "Matous", "surname": "Holinka", "email": "matous@nic.cz"}), # dict
    sep="\n"
)
````

````python
"Matous".__len__()
````

<br>

You see that the `len` function can work with different data types.

<br>

It returns a specific value for each data type separately. But you're still using one and the same function.
1. `str`, length of the string,
2. `list`, number of items in the sequence,
3. `dict`, number of keys in the object

<br>

### Usage in OOP

---

Furthermore, this concept allows overriding names and usage of **methods**:

````python
class CsvProcessor:
    def read_file(self) -> str:
        return "Opening CSV file..."
````

````python
class JsonProcessor:
    def read_file(self) -> str:
        return "Opening JSON file..."
````

````python
csv_file = CsvProcessor()
json_file = JsonProcessor()
````

````python
for file in (csv_file, json_file):
    print(file.read_file())
````

<br>

<br>

Practically, this means that *polymorphism* allows objects of **different classes** to work with **methods of the same name**.

You can encounter further use of *polymorphism* in inheritance (later in the materials).

## Encapsulation

---

### Terminology

---

This term generally refers to **some form of hiding**.

It's primarily about working **with private and protected** attributes or methods.

The purpose of this hiding, or encapsulation, is protection.

<br>

To prevent programmers from overwriting what they shouldn't and thereby causing **complications** in the source code.

### Without the concept of encapsulation

---

In combination with **protected objects**, you want to secure certain processes while not complicating user usage:

````python
class MyApplication:
    
    def __init__(self, name: str, credit: int):
        self.name = name
        self.credit = credit
        
    def display_status(self):
        return f"User: {self.name}, available credits: {self.credit}"
````

````python
user_matous = MyApplication("Matous", 100)
````

````python
print(user_matous.display_status())
````

<br>

In the current state of the class, **anyone from outside** our `MyApplication` class can accidentally overwrite values:

````python
user_matous.credit = 10_000
````

````python
print(user_matous.display_status())
````

````python
user_matous.credit = "10_000"
````

````python
print(user_matous.display_status())
````

````python
user_matous.credit += 5_000
````

<br>

### Private objects

---

<br>

Therefore, it's advisable to supplement the code with:
1. **protected** `__name`,
2. **private** `_name`.

````python
class MyApplication:
    
    def __init__(self, name: str, credit: int):
        self.name = name
        self.__credit = credit
        
    def display_status(self):
        return f"User: {self.name}, available credits: {self.__credit}"
````

````python
user_lukas = MyApplication("Lukas", 1000)
````

````python
print(user_lukas.display_status())
````

````python
user_lukas.credit = 1_000_000
````

````python
print(user_lukas.display_status())
````

````python
user_lukas.__dict__
````

<br>

The concept on which a private object in Python is based is called name mangling (that is, distorting object names).

However, this kind of measure still does not solve the problem if an instance receives, for example, an inappropriate data type:

````python
user_lukas._MyApplication__credit = "1_000_000"
````

````python
print(user_lukas.display_status())
````

### Getters, setters

---

<br>

For complete order, it's necessary to establish the form in which the attribute will be entered:

````python
class MyApplication:
    
    def __init__(self, name: str):
        self.name = name
        self.__credit = 0
        
    @property
    def credit(self):
        """Getter method"""
        return self.__credit
    
    @credit.setter
    def credit(self, value: int) -> None:
        """Setter method"""
        if isinstance(value, int):
            self.__credit = value
        else:
            raise Exception("Entered value is not an integer")
        
    def display_status(self):
        return f"User: {self.name}, available credits: {self.__credit}"
````

````python
user_lukas = MyApplication("Lukas")
````

````python
user_lukas.credit = 1000
````

````python
print(user_lukas.display_status())
````

<br>

Now it's not possible to enter a data type other than an integer:

````python
user_lukas.credit = "1000"
````

<br>

Using the default `Exception`, the user now sees that it's not possible to work with a different data type.

#### Summary of encapsulation
* **security**, private objects have an additional layer of protection against unintended overwriting or using an incorrect data type,
* **control**, other programmers must follow procedures for manipulating private objects,
* **simplicity**, separation of setting and getting values are separate objects,
* **readability**, associating objects with their respective classes makes the code more clear.

### 🧠 EXERCISE 8 🧠, Create a class `TeslaCar` and add the following:

- Define a class `TeslaCar`, with a method `__init__` that only needs parameters `user` and `number`,
- create **getter and setter** methods for the attribute `max_speed`,
- limit entering `max_speed` only for data type `int` from the range `0 - 320`,
- if the user doesn't meet the conditions, raise an exception with text: `Unacceptable data type or value (< 320 km/h)`,
- create **getter and setter** methods for the attribute `max_range`,
- limit entering `max_range` only for data type `int` from the range `0 - 590`,
- if the user doesn't meet the conditions, raise an exception with text: `Unacceptable data type or value (< 600 km)`,
- override the magic method `__str__` to format this output: `User: <user>, serial number: <number>`.

````python
class TeslaCar():
    def __init__(self, user: str, number: int):
        self.user = user
        self.number = number

    @property
    def max_speed(self):
        return self.__max_speed
    
    @max_speed.setter
    def max_speed(self, value: int):
        if value < 320 and isinstance(value, int):
            self.__max_speed = value
        else:
            raise ValueError("Max speed cannot exceed 320 km/h.")
        
    @property
    def max_range(self):
        return self.__max_range
    
    @max_range.setter
    def max_range(self, value: int):
        if value < 590 and isinstance(value, int):
            self.__max_range = value
        else:
            raise ValueError("Max range cannot exceed 590 km.")
        
    def __str__(self):
        return f"User {self.user}, serial number: {self.number}"
````
