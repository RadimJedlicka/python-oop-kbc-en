### Why Virtual Environments

---

Imagine you're a chef who cooks both Italian and Japanese cuisine.

Would you store all your ingredients in the same pantry without any organization?

<br>

Probably not! You'd want to keep them separate to avoid confusion and ensure each dish gets exactly what it needs.

**Virtual environments** work the same way for your Python projects.

<br>

### The Global Installation Problem

---

When you install Python packages **globally** (system-wide), you can run into several problems:

1. **Version conflicts**: Project A needs `requests==2.25.0`, but Project B needs `requests==2.31.0`
2. **Dependency hell**: Installing a package for one project might break another
3. **Reproducibility issues**: Hard to know exactly which packages a project needs
4. **System pollution**: Your global Python installation becomes cluttered

<br>

Let's see what happens with global installations:

````python
# BAD: Global installation scenario (pseudocode)

# Project A needs:
# pip install requests==2.25.0
# pip install pandas==1.2.0

# Project B needs:
# pip install requests==2.31.0  # This overwrites the version Project A needs!
# pip install pandas==2.0.0     # This breaks Project A!

# Result: Project A is now broken because its dependencies were overwritten
````

<br>

### How Virtual Environments Solve This

---

A **virtual environment** is an isolated Python installation that:

- Has its **own set of packages**, separate from other environments
- Can have **different package versions** than other environments
- Lives in a **specific directory** within your project
- Can be **easily deleted and recreated**

<br>

Think of it as creating a **separate kitchen** for each cuisine you cook!

````python
# EXCELLENT: Virtual environment scenario (pseudocode)

# Project A (in its own virtual environment):
# Has requests==2.25.0 and pandas==1.2.0

# Project B (in its own virtual environment):
# Has requests==2.31.0 and pandas==2.0.0

# Result: Both projects work perfectly with their required versions!
````

<br>

## venv - The Standard Library Solution

---

Python comes with a built-in module called `venv` for creating virtual environments.

It's part of the **standard library** since Python 3.3, so you don't need to install anything extra!

<br>

There's also an older tool called `virtualenv` that works similarly, but for most cases, `venv` is all you need.

<br>

| Tool | Description | When to Use |
|------|-------------|-------------|
| `venv` | Built-in Python module | Default choice for most projects |
| `virtualenv` | Third-party package | When you need extra features or Python 2 support |

<br>

### Creating a Virtual Environment

---

To create a virtual environment, use the `venv` module with the following command:

<br>

**Specify Python version (or use default):**
```bash
pyenv local 3.12
# ..and check the selected version
```


**Linux/macOS:**
```bash
python3 -m venv .venv
```

**Windows:**
```powershell
python -m venv .venv
```

<br>

Let's break down this command:

| Part | Meaning |
|------|--------|
| `python3` / `python` | The Python interpreter to use |
| `-m venv` | Run the `venv` module as a script |
| `.venv` | The name/path of the directory to create |

<br>

After running this command, you'll see a new directory called `.venv` with the following structure:

```
.venv/
├── bin/          # (Linux/macOS) or Scripts/ (Windows)
│   ├── activate  # Activation script
│   ├── python    # Python interpreter symlink
│   └── pip       # pip for this environment
├── include/      # C headers (for compiled packages)
├── lib/          # Installed packages go here
└── pyvenv.cfg    # Configuration file
```

<br>

### Activation and Deactivation

---

Creating a virtual environment is just the first step. To **use** it, you need to **activate** it.

<br>

**Activation** tells your shell to use the Python interpreter and packages from the virtual environment instead of the global ones.

<br>

**Linux/macOS (bash/zsh):**
```bash
source .venv/bin/activate
```

**Windows (Command Prompt):**
```cmd
.venv\Scripts\activate.bat
```

**Windows (PowerShell):**
```powershell
.venv\Scripts\Activate.ps1
```

<br>

Once activated, you'll notice your **terminal prompt changes** to show the environment name:

```
# Before activation:
user@machine:~/project$

# After activation:
(.venv) user@machine:~/project$
```

<br>

You can verify which Python interpreter is being used:

**Linux/macOS:**
```bash
which python
# Output: /path/to/project/.venv/bin/python
```

**Windows:**
```powershell
where python
# Output: C:\path\to\project\.venv\Scripts\python.exe
```

<br>

To **deactivate** the virtual environment and return to the global Python:

```bash
deactivate
```

This command works the same on all operating systems.

<br>

## Best Practices

---

### Naming Conventions

---

The most common names for virtual environment directories are:

| Name | Description |
|------|-------------|
| `.venv` | **Recommended** - hidden directory, clear purpose |
| `venv` | Common, but visible in file listings |
| `.env` | Avoid - can be confused with environment variable files |
| `env` | Avoid - too generic, can be confused with other tools |

<br>

We recommend using **`.venv`** because:
- The leading dot makes it **hidden** on Linux/macOS
- It's **clearly named** so everyone knows what it is
- It's the **convention** used by most Python developers

<br>

### Location and .gitignore

---

**Where to create the virtual environment?**

Always create it **inside your project directory**:

```
my_project/
├── .venv/          # Virtual environment lives here
├── src/
│   └── main.py
├── requirements.txt
└── README.md
```

<br>

**Never commit your virtual environment to Git!**

Virtual environments contain:
- Platform-specific binaries
- Thousands of files
- Absolute paths to your system

Instead, add it to your `.gitignore` file:

```gitignore
# Virtual environments
.venv/
venv/
ENV/
```

<br>

To share your project's dependencies with others, use a **requirements.txt** file:

```bash
# Save current packages to requirements.txt
pip freeze > requirements.txt

# Install packages from requirements.txt (on another machine)
pip install -r requirements.txt
```

This way, anyone can recreate your environment by running `pip install -r requirements.txt`.

<br>

## Complete Workflow Example

---

Here's a complete workflow for starting a new Python project with a virtual environment:

<br>

**Linux/macOS:**
```bash
# 1. Create project directory
mkdir my_project
cd my_project

# 2. Create virtual environment
python3 -m venv .venv

# 3. Activate it
source .venv/bin/activate

# 4. Install packages
pip install requests pandas

# 5. Save dependencies
pip freeze > requirements.txt

# 6. Work on your project...

# 7. Deactivate when done
deactivate
```

<br>

**Windows (PowerShell):**
```powershell
# 1. Create project directory
mkdir my_project
cd my_project

# 2. Create virtual environment
python -m venv .venv

# 3. Activate it
.venv\Scripts\Activate.ps1

# 4. Install packages
pip install requests pandas

# 5. Save dependencies
pip freeze > requirements.txt

# 6. Work on your project...

# 7. Deactivate when done
deactivate
```

| **Why use it** | Avoid version conflicts, keep projects independent |
| **venv** | Built-in module, `python -m venv .venv` |
| **Activation** | `source .venv/bin/activate` (Linux/macOS) or `.venv\Scripts\activate` (Windows) |
| **Deactivation** | `deactivate` |

### 🧠 EXERCISE 🧠, Recreate an Environment from requirements.txt

---

Simulate what happens when a colleague clones your project:

1. Delete your `.venv` directory (simulating a fresh clone)
2. Create a new virtual environment
3. Activate it
4. Install all packages from `requirements.txt`
5. Verify that `requests` is installed

<details>
    <summary>▶️ Solution (Linux/macOS)</summary>
    
```bash
# 1. Delete old environment
rm -rf .venv

# 2. Create new environment
python3 -m venv .venv

# 3. Activate
source .venv/bin/activate

# 4. Install from requirements.txt
pip install -r requirements.txt

# 5. Verify requests is installed
pip list | grep requests
# Expected: requests    2.x.x
```
</details>

<details>
    <summary>▶️ Solution (Windows PowerShell)</summary>
    
```powershell
# 1. Delete old environment
Remove-Item -Recurse -Force .venv

# 2. Create new environment
python -m venv .venv

# 3. Activate
.venv\Scripts\Activate.ps1

# 4. Install from requirements.txt
pip install -r requirements.txt

# 5. Verify requests is installed
pip list | Select-String requests
# Expected: requests    2.x.x
```
