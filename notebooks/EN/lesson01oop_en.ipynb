### Class attribute

A **class attribute** or also **class variable** is a variable (~property) that belongs to the class itself.

Because the *class attribute* belongs to the class, I can access it **without creating an instance**.

````python
class Employee:
    """Class object representing an employee template."""
    max_vacation_days = 20
````

````python
matous = Employee()
petr   = Employee()
````

````python
print(matous.max_vacation_days)
````

````python
print(petr.max_vacation_days)
````

<br>

Without access to the *instance* itself:

````python
print(Employee.max_vacation_days)
````

If the *class attribute* is **immutable**, you can overwrite its value **ONLY within an instance**.

````python
matous.max_vacation_days = 25
````

````python
print(
    matous.max_vacation_days,
    petr.max_vacation_days,
    Employee.max_vacation_days,
    sep="\n"
)
````

Notice that modifying the value within the `matous` instance has no effect on either the original class or the `petr` instance.

Only if you overwrite the original value directly **in the class definition**, then you change the value of the *class attribute* also for **class instances**.

````python
Employee.max_vacation_days = 30
````

````python
print(Employee.max_vacation_days)
````

````python
print(
    matous.max_vacation_days,
    petr.max_vacation_days,
    sep='\n'
)
````

````python
matous = Employee()
````

````python
print(
    matous.max_vacation_days,
    petr.max_vacation_days,
    Employee.max_vacation_days,
    sep="\n"
)
````

When you create a *class attribute* that **is mutable**, be careful how it works:

````python
class Employee:
    """Class object representing an employee template."""
    benefits = list()
````

````python
matous = Employee()
petr = Employee()
````

````python
print(
    matous.benefits,
    petr.benefits,
    Employee.benefits,
    sep="\n"
)
````

..as soon as you modify the *class attribute* within **a single instance** this time..

````python
matous.benefits.append("Sodexo")
````

````python
print(
    matous.benefits,
    petr.benefits,
    Employee.benefits,
    sep="\n"
)
````

..you get a changed value within **all instances** AND ALSO **in the class itself**.

### Summary

So if you work with *class attributes*, as:
1. **immutable** data type, overwriting within an *instance* **does not affect the original class** (or other *instances*),
2. **mutable** data type, overwriting within an *instance* **affects the original class** (or other *instances*).

In Python, the second type of attributes is more commonly used, namely **instance attributes**.

*Class attributes* usually serve as an opportunity to write a common constant, helper object within the class.

### Instance attribute

---

An **instance attribute**, on the other hand, is a variable that belongs **only to the created instance**.

The class itself has no access to it.

<br>

#### Creating a class instance

So far you have been creating *instances* that are **only empty**:

````python
class Employee:
    """Class object representing an employee template."""
    benefits = list()
````

Because you don't use the **instance constructor**, `__init__`, you create an instance using the so-called **empty constructor**.

However, if you need each instance to carry different attributes, you will have to use the **special instance constructor**, the `__init__` method.

In general, Python uses two special methods to create an instance.

The methods that the interpreter needs to **create a new instance** are:
1. `__new__`, *creates* a new object (no need to define),
2. `__init__`, *initializes* a new object.

````python
class Employee:
    """Class object representing an employee template."""

    def __init__(self, name, last_name):  # constructor for new instances
        self.name = name                # 1. instance attribute
        self.last_name = last_name          # 2. instance attribute
````

Notice that the `__init__` constructor is called a *method*, but still looks like a regular user function.

In each `__init__` method, attributes are defined that will be individual for potential instances.

<br>

If you know that employee instances will contain a first name and last name, you must show them to this method.

You put them in the parameters and the `__init__` method will add them to the *instance*.

````python
print(
    Employee("Matouš", "Holinka"),
    Employee("Petr", "Svetr"),
    sep="\n"
)
````

````python
matous = Employee("Matouš", "Holinka")
petr   = Employee("Petr", "Svetr")
````

````python
print(matous, petr, sep="\n")
````

````python
print(
    matous.name,
    matous.last_name,
    sep="\n"
)
````

````python
print(
    petr.name,
    petr.last_name,
    sep="\n"
)
````

````python
def func(x, y, z):
    pass
````

````python
func(1, 2)
````

<br>

So the `__init__` method in the class definition needs the parameters `self`, `name`, `last_name`.

But to create a **new instance** of the `Employee` class, it was enough to write **only two arguments**.

<br>

What then does the *parameter* `self` represent?

`self` is basically a **helper reference** that helps the *interpreter* direct at the moment of instance creation:

````python
class Employee:
    """Class object representing an employee template."""

    def __init__(self, name, last_name):  # constructor for new instances
        self.name = name                # 1. instance attribute
        self.last_name = last_name          # 2. instance attribute
````

````python
matous = Employee("Matouš", "Holinka")  # valid notation for a new instance,
                                           # "matous" is the instance name and you omit the argument "self"
````

````python
print(matous.name)
````

````python
# The example below is only PSEUDOCODE, for better understanding,
Employee(matous, "Matouš", "Holinka")   # pseudocode, where instead of "self"
                                           # ..I specify the instance name, self = "Matouš"
````

`self` is thus prepared in your definition, in the `Employee` class and waits until you want to create a new instance.

<br>

A few details about `self`:
1. Deliberately written as the **first parameter**,
2. in other languages it is also referred to by other words, e.g. *Java* uses the expression `this`,
3. part of all methods where you want to **make instance attributes accessible**,
4. you can use a different expression than `self`, but `self` is used as a convention among other programmers.

### Class method

---

Classes don't necessarily have to have only properties.

They can also do something, i.e., have certain **capabilities**.

The *capabilities* of a class are called **methods**:

````python
class Employee:
    """Class object representing an employee template."""

    def __init__(self, name, last_name):  # constructor for new instances
        self.name = name                # 1. instance attribute
        self.last_name = last_name          # 2. instance attribute
        self.email = ""
        
    def create_email(self, domain: str = "superfirma.cz"):
        # return f"{self.name[0].lower()}.{self.last_name.lower()}@{domain}"
        self.email = f"{self.name[0].lower()}.{self.last_name.lower()}@{domain}"
````

At first glance, you can say that it doesn't differ much from a regular **user function**.

It is basically indeed a function that falls into the class definition.

Its use then depends on the created instance:

````python
matous = Employee("Matouš", "Holinka")
````

````python
print(matous.email)
````

````python
new_email = matous.create_email()
````

````python
print(new_email)
````

````python
petr = Employee("Petr", "Svetr")
````

````python
petr.create_email()
````

````python
print(petr.__dict__)
````

````python
class Employee:
    """Class object representing an employee template."""

    def __init__(self, name, last_name):  # constructor for new instances
        self.name = name                # 1. instance attribute
        self.last_name = last_name          # 2. instance attribute
        self.email = self.create_email()
        
    def create_email(self, domain: str = "superfirma.cz"):
        return f"{self.name[0].lower()}.{self.last_name.lower()}@{domain}"
````

    total_accounts = 0

    def __init__(self, account_number):
        self.account_number = account_number
        self.balance = 0
        BankAccount.total_accounts += 1

    def deposit(self, amount):
        self.balance += amount

        def __init__(self, priority: range):
            self.priority = priority
            self.tickets: list = []

        def create_ticket(self, user: str, issue: str) -> None:
            self.tickets.append(TicketGenerator(user, issue))

        def process_ticket(self) -> None:
            if not self.tickets:
                print("There are no unresolved tickets, well done..")
            else:
                for ticket in self.tickets:
                    importance: int = random.choice(self.priority) # add specific value as parameter
                    self.show_status(ticket, importance)

        def show_status(self, ticket: TicketGenerator, priority: int) -> None:
            print(
                f"Created: {ticket.date}",
                f"Processing ticket id: {ticket.uuid}",
                f"Issue: {ticket.issue}",
                f"Customer: {ticket.user}",
                f"Importance: {priority}",
                "=" * 58,
                sep="\n"
            )
