### Class attribute

A **class attribute** or also **class variable** is a variable (~property) that belongs to the class itself.

Because the *class attribute* belongs to the class, I can access it **without creating an instance**.

````python
class Employee:
    """Class object representing an employee template."""
    max_vacation_days = 20
````

````python
matous = Employee()
petr   = Employee()
````

````python
print(matous.max_vacation_days)
````

````python
print(petr.max_vacation_days)
````

<br>

Without access to the *instance* itself:

````python
print(Employee.max_vacation_days)
````

If the *class attribute* is **immutable**, you can overwrite its value **ONLY within an instance**.

````python
matous.max_vacation_days = 25
````

````python
print(
    matous.max_vacation_days,
    petr.max_vacation_days,
    Employee.max_vacation_days,
    sep="\n"
)
````

Notice that modifying the value within the `matous` instance has no effect on either the original class or the `petr` instance.

Only if you overwrite the original value directly **in the class definition**, then you change the value of the *class attribute* also for **class instances**.

````python
Employee.max_vacation_days = 30
````

````python
print(Employee.max_vacation_days)
````

````python
print(
    matous.max_vacation_days,
    petr.max_vacation_days,
    sep='\n'
)
````

````python
matous = Employee()
````

````python
print(
    matous.max_vacation_days,
    petr.max_vacation_days,
    Employee.max_vacation_days,
    sep="\n"
)
````

When you create a *class attribute* that **is mutable**, be careful how it works:

````python
class Employee:
    """Class object representing an employee template."""
    benefits = list()
````

````python
matous = Employee()
petr = Employee()
````

````python
print(
    matous.benefits,
    petr.benefits,
    Employee.benefits,
    sep="\n"
)
````

..as soon as you modify the *class attribute* within **a single instance** this time..

````python
matous.benefits.append("Sodexo")
````

````python
print(
    matous.benefits,
    petr.benefits,
    Employee.benefits,
    sep="\n"
)
````

..you get a changed value within **all instances** AND ALSO **in the class itself**.

### Summary

So if you work with *class attributes*, as:
1. **immutable** data type, overwriting within an *instance* **does not affect the original class** (or other *instances*),
2. **mutable** data type, overwriting within an *instance* **affects the original class** (or other *instances*).

In Python, the second type of attributes is more commonly used, namely **instance attributes**.

*Class attributes* usually serve as an opportunity to write a common constant, helper object within the class.

### Instance attribute

---

An **instance attribute**, on the other hand, is a variable that belongs **only to the created instance**.

The class itself has no access to it.

#### Creating a class instance

So far you have been creating *instances* that are **only empty**:

````python
class Employee:
    """Class object representing an employee template."""
    benefits = list()
````

Because you don't use the **instance constructor**, `__init__`, you create an instance using the so-called **empty constructor**.

However, if you need each instance to carry different attributes, you will have to use the **special instance constructor**, the `__init__` method.

In general, Python uses two special methods to create an instance.

The methods that the interpreter needs to **create a new instance** are:
1. `__new__`, *creates* a new object (no need to define),
2. `__init__`, *initializes* a new object.

````python
class Employee:
    """Class object representing an employee template."""

    def __init__(self, name, last_name):  # constructor for new instances
        self.name = name                # 1. instance attribute
        self.last_name = last_name          # 2. instance attribute
````

Notice that the `__init__` constructor is called a *method*, but still looks like a regular user function.

In each `__init__` method, attributes are defined that will be individual for potential instances.

<br>

If you know that employee instances will contain a first name and last name, you must show them to this method.

You put them in the parameters and the `__init__` method will add them to the *instance*.

````python
print(
    Employee("Matouš", "Holinka"),
    Employee("Petr", "Svetr"),
    sep="\n"
)
````

````python
matous = Employee("Matouš", "Holinka")
petr   = Employee("Petr", "Svetr")
````

````python
print(matous, petr, sep="\n")
````

````python
print(
    matous.name,
    matous.last_name,
    sep="\n"
)
````

````python
print(
    petr.name,
    petr.last_name,
    sep="\n"
)
````

### 🧠 EXERCISE 🧠 Create a `BankAccount` class that has the following **attributes and methods**
---
- **`total_accounts` (class attribute):** the total number of created accounts (starts at 0)
- **`account_number` (instance attribute):** a unique account number (e.g., generated automatically or provided by the user)
- **`balance` (instance attribute):** the current account balance (starts at 0)
- **`deposit(amount)` method:** deposits the specified amount of money into the account and updates the balance
- **`withdraw(amount)` method:** withdraws the specified amount of money from the account and updates the balance (if there are sufficient funds in the account)
- **`get_balance()` method:** returns the current account balance with a timestamp
- **`get_total_accounts()` method:** returns the total number of created accounts
- When a new `BankAccount` instance is created, automatically increment **`total_accounts`** by 1
- Create several instances of the `BankAccount` class with different account numbers and perform several deposits and withdrawals
- Display the balance of each account and the total number of created accounts

````python
class BankAccount:
    total_accounts = 0

    def __init__(self, account_number: str):
        self.account_number = account_number
        BankAccount.total_accounts += 1
        self.balance = 0.0

    def deposit(self, amount: float) -> None:
        self.balance += amount

    def withdraw(self, amount: float) -> None:
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
    
    def get_balance(self) -> float:
        return self.balance
    
    def get_total_accounts(self) -> int:
        return BankAccount.total_accounts
````

````python
acc1 = BankAccount("123456789")
acc2 = BankAccount("987654321")
````

````python
acc1.deposit(1000)
acc2.deposit(500)
acc1.withdraw(200)
acc2.withdraw(100)
````

````python
print(acc1.get_balance())
print(acc2.get_balance())
````

### Vysvětlení pro studenty: Proč nefunguje `print("Total created accounts:", BankAccount.get_total_accounts())`?

#### Česky
V kódu máte metodu `get_total_accounts` definovanou jako **instance metodu** (bez dekorátoru `@classmethod`):

```python
def get_total_accounts(self) -> int:
    return BankAccount.total_accounts
```

Když se pokusíte volat `BankAccount.get_total_accounts()`, dostanete chybu `TypeError: get_total_accounts() missing 1 required positional argument: 'self'`. Proč?

- **Instance metody** potřebují jako první argument instanci (objekt), který se obvykle nazývá `self`. Volání na třídě bez instance selže, protože Python neví, co má předat jako `self`.
- Metoda je stále instance metoda, i když nepoužívá `self` uvnitř – první parametr je stále `self`.
- Aby to fungovalo, musíte přidat dekorátor `@classmethod` a změnit `self` na `cls`:

```python
@classmethod
def get_total_accounts(cls) -> int:
    return BankAccount.total_accounts
```

Pak můžete volat `BankAccount.get_total_accounts()` přímo na třídě.

#### English
In the code, you have the `get_total_accounts` method defined as an **instance method** (without the `@classmethod` decorator):

```python
def get_total_accounts(self) -> int:
    return BankAccount.total_accounts
```

When you try to call `BankAccount.get_total_accounts()`, you get a `TypeError: get_total_accounts() missing 1 required positional argument: 'self'`. Why?

- **Instance methods** require an instance (object) as the first argument, usually called `self`. Calling on the class without an instance fails because Python doesn't know what to pass as `self`.
- The method is still an instance method, even if it doesn't use `self` inside – the first parameter is still `self`.
- To make it work, you need to add the `@classmethod` decorator and change `self` to `cls`:

```python
@classmethod
def get_total_accounts(cls) -> int:
    return BankAccount.total_accounts
```

Then you can call `BankAccount.get_total_accounts()` directly on the class.

````python
print(acc1.get_total_accounts())
````

````python
print(BankAccount.total_accounts)
````

        self.balance = 0.0
        BankAccount.total_accounts += 1

    def deposit(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError('Deposit amount must be positive')
        self.balance += amount

    def withdraw(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError('Withdrawal amount must be positive')
        if self.balance >= amount:
            self.balance -= amount
        else:
            print('Insufficient funds')

    def get_balance(self) -> str:
        ts = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        return f'Account {self.account_number} balance: {self.balance:.2f} (as of {ts})'

    def get_total_accounts(self) -> int:
        return BankAccount.total_accounts
